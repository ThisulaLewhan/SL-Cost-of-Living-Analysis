```python
df_usd = pd.read_excel('Exchange_Rates.xlsx', header=5, skiprows=[6,7])
print(df_usd.head())
```

```python
df_usd['Year'] = df_usd['Year'].ffill()
print(df_usd.head(10))
```

```python
df_usd = df_usd.dropna(subset=['Month']).copy()
df_usd['Date'] = pd.to_datetime(df_usd['Year'].astype(str) + '-' + df_usd['Month'].astype(str) + '-01')
print(df_usd.head())
```

```python
df_usd = df_usd[['Date', 'USD']]
df_usd = df_usd.sort_values(by='Date')
print(df_usd.head())
```

```python
df_usd = df_usd[df_usd['Date'] >= '2020-01-01']
print(df_usd.head())
```

```python
df_fuel = pd.read_excel('Fuel_Prices.xlsx')
print(df_fuel.head())
```

```python
df_fuel['Date'] = pd.to_datetime(df_fuel['Date'])
df_fuel = df_fuel.sort_values(by='Date')
print(df_fuel.head())
```

```python
df_fuel.set_index('Date', inplace=True)
print(df_fuel.head())
```

```python
df_fuel_monthly = df_fuel.resample('ME').ffill().reset_index()
print(df_fuel_monthly.head())
```

```python
df_fuel_monthly = df_fuel_monthly.rename(columns={'Petrol': 'Petrol_LKR', 'Diesel': 'Diesel_LKR'})
print(df_fuel_monthly.head())
```

```python
df_temp = pd.merge(df_cpi, df_usd, on='Date', how='left')
print(df_temp.head())
```

```python
df_master = pd.merge(df_temp, df_fuel_monthly, on='Date', how='left')
print(df_master.info())
```

```python
df_cpi['Date'] = df_cpi['Date'].dt.replace(day=1)
```

```python
df_cpi['Date'] = df_cpi['Date'].apply(lambda x: x.replace(day=1))
print(df_cpi.head())
```

```python
df_fuel_monthly = df_fuel.resample('MS').ffill().reset_index()
df_fuel_monthly = df_fuel_monthly.rename(columns={'Petrol': 'Petrol_LKR', 'Diesel': 'Diesel_LKR'})
print(df_fuel_monthly.head())
```

```python
df_usd_monthly = df_usd.resample('MS', on='Date').mean().reset_index()
print(df_usd_monthly.head())
```

```python
df_temp = pd.merge(df_cpi, df_usd_monthly, on='Date', how='left')
print(df_temp.head())
```

```python
df_master = pd.merge(df_temp, df_fuel_monthly, on='Date', how='left')
print(df_master.info())
```
